```python
path_prefix = '../..'

import sys
sys.path.append(path_prefix)

from sklearn.metrics import accuracy_score, f1_score, precision_recall_curve, auc
from egh_vlm.generate_answers import generate_enhanced_answers
```

## Analysis

```python
res = generate_enhanced_answers(
    dataset_path=f'{path_prefix}/data/phd/prototype/phd_sample_qwen3_vl_2b_scored.json',
    img_folder_path=f'{path_prefix}/data/phd/images',
    prompt_path=f'{path_prefix}/data/phd/enhance_v0.txt',
    save_path=f'{path_prefix}/data/phd/prototype/phd_sample_qwen3_vl_2b_enhanced.json',
)
```

```python
labels = [s['label'] for s in res]
preds_base = [s['baseline_pred'] for s in res]
preds_enhanced = [s['enhanced_pred'] for s in res]
```

```python
print("Base Prediction:")
print(f"F1: {f1_score(labels, preds_base)}")
print(f"ACC: {accuracy_score(labels, preds_base)}")
precision, recall, cm = precision_recall_curve(labels, preds_base)
print(f"PR AUC: {auc(recall, precision)}")
```

```python
print("Enhanced Prediction:")
print(f"F1: {f1_score(labels, preds_enhanced)}")
print(f"ACC: {accuracy_score(labels, preds_enhanced)}")
precision, recall, cm = precision_recall_curve(labels, preds_enhanced)
print(f"PR AUC: {auc(recall, precision)}")
```
